```python
# Set up the API keys
from google.colab import userdata

# Retrieve the API key
NYTIMES_API_KEY = userdata.get('NYTIMES_API_KEY')

if NYTIMES_API_KEY:
    print("NYTimes API key retrieved successfully!")
else:
    print("Failed to retrieve NYTimes API key. Please check your setup.")

# Retrieve the API key
TMDB_API_KEY = userdata.get('TMDB_API_KEY')

# Get API key
if TMDB_API_KEY:
    print("TMDB API key retrieved successfully!")
else:
    print("Failed to retrieve TMDB API key. Please check your setup.")
```

```python
import requests

# Getting actor name from API
def get_actor_name(ticker):
  url = f'https://api.themoviedb.org/3/search/person?api_key={TMDB_API_KEY}&search_type=ngram&query={ticker}'
  response = requests.get(url)
  data = response.json()

  if 'results' in data:
    return data['results'][0]['name']
  else:
    print('Name not found')
    return None

#test
ticker = "Taylor"
actor_name = get_actor_name(ticker)
print(actor_name)
```

```python
# Get well known movies for actor on the API
def get_actor_movies(actor_name):
  url = f'https://api.themoviedb.org/3/search/person?api_key={TMDB_API_KEY}&search_type=ngram&query={actor_name}'
  response = requests.get(url)

  if response.status_code == 200:
    data = response.json()
    return data
  else:
    print('Failed to retrieve movies')
    return None
```

```python
# Return well known movies for actor on the API
def get_movies_for_ticker(ticker):
  actor_name = get_actor_name(ticker)
  print("Actor(ress): ", actor_name)
  if actor_name:
    movies = get_actor_movies(actor_name)
    return movies['results'][0]["known_for"]
  else:
    return []

# Display well known movies for actor from the API
def display_movies(movies):
  if movies:
    print(f'Found {len(movies)} movies')
    for movie in movies[:10]:
      print(f"Title: {movie['title']}")
      print(f"Overview: {movie['overview']}\n")
  else:
    print("No movies found")

ticker = input("Enter an actor name ticker symbol: ")
movies = get_movies_for_ticker(ticker)
display_movies(movies)
```

```python
# Using actor name from TMDB API to get news from NYTimes API
def search_news(actor_name):
  url = f'https://api.nytimes.com/svc/search/v2/articlesearch.json?q={actor_name}&api-key={NYTIMES_API_KEY}'
  response = requests.get(url)

  if response.status_code == 200:
    data = response.json()
    return data
  else:
    print('Failed to retrieve news')
    return None

# print(search_news("Taylor"))
```

```python
# Get actor name from TMDB API to get news from NYTimes API
def get_news_for_ticker(ticker):
  actor_name = get_actor_name(ticker)
  if actor_name:
    news = search_news(actor_name)
    return news['response']['docs']
  else:
    return []

# Display actor name from TMDB API to get news from NYTimes API
def display_news(news):
  if news:
    print(f'Found {len(news)} news')
    for new in news[:10]:
      print(f"Title: {new['headline']['main']}")
      print(f"URL: {new['web_url']}\n")
  else:
    print("No news found")

ticker = input("Enter an actor name ticker symbol: ")
news = get_news_for_ticker(ticker)
display_news(news)
```
